```python
#  loại bỏ các ký tự như dấu phẩy, ký hiệu đô la và ký hiệu phần trăm khỏi các giá trị trong các cột
# Columns to convert to float
columns_to_convert = ['Density\n(P/Km2)', 'Agricultural Land( %)', 'Land Area(Km2)',
                      'Birth Rate', 'Co2-Emissions', 'Forested Area (%)',
                      'CPI', 'CPI Change (%)', 'Fertility Rate', 'Gasoline Price', 'GDP',
                      'Gross primary education enrollment (%)', 'Armed Forces size',
                      'Gross tertiary education enrollment (%)', 'Infant mortality',
                      'Life expectancy', 'Maternal mortality ratio', 'Minimum wage',
                      'Out of pocket health expenditure', 'Physicians per thousand',
                      'Population', 'Population: Labor force participation (%)',
                      'Tax revenue (%)', 'Total tax rate', 'Unemployment rate', 'Urban_population']

# Convert columns using a lambda function
df[columns_to_convert] = df[columns_to_convert].applymap(lambda x: float(str(x).replace(',', '').replace('$', '').replace('%', '')))
```

top 20 nước có tỷ lệ thất nghiệp cao nhất (3)

20 quốc gia có tỷ lệ thất nghiệp cao nhất: Gabon có tỷ lệ thất nghiệp cao nhất ở mức 20%, tiếp theo là Botswana và Hy Lạp. Mỹ cũng góp mặt trong danh sách này với tỷ lệ thất nghiệp khá cao.

```python
sorted_df = df.sort_values(by = 'Unemployment rate', ascending = False)

top_high_unemployment = sorted_df.head(20)




# Biểu đò cho các quốc gia thất nghiệp cao
ax = sns.barplot(data=top_high_unemployment, x='Country', y='Unemployment rate', palette='bright')
plt.title('Top 20 quốc gia có tỷ lệ thất nghiệp cao')
plt.xlabel('Quốc gia')
plt.ylabel('Tỷ lệ thất nghiệp')
plt.xticks(rotation=45, ha='right')

for p in ax.patches:
    plt.annotate(f'{p.get_height():.2f}%',
                 (p.get_x() + p.get_width() / 2, p.get_height()),
                 ha='center', va='bottom', fontsize=10, color='black',
                 xytext=(0, 5), textcoords='offset points', rotation=45)

# Hiển thị đồ thị
plt.show()
```

```python
# Sắp xếp DataFrame theo cột 'Tỷ lệ tuyển sinh giáo dục đại học (Gross)' theo thứ tự giảm dần và lấy 20 hàng đầu
top_tertiary_enrollment_countries = df.sort_values(by='Gross tertiary education enrollment (%)', ascending=False).head(20)

# Tạo biểu đồ cột để hiển thị tỷ lệ nhập học đại học cho các quốc gia hàng đầu
ax = sns.barplot(data=top_tertiary_enrollment_countries, x='Gross tertiary education enrollment (%)', y='Country', palette='bright')
plt.title('Top 20 Quốc gia có Tỷ lệ nhập học đại học cao nhất')
plt.xlabel('Tỷ lệ nhập học đại học (%)')
plt.ylabel('Quốc gia')
plt.xticks(rotation=45, ha='right')

# Ghi chú giá trị chính xác của từng cột
for index, value in enumerate(top_tertiary_enrollment_countries['Gross tertiary education enrollment (%)']):
    ax.text(value + 1, index, f'{value:.0f}', va='center', color='black')

plt.tight_layout()
plt.show()
```

14| Các quốc gia theo ngôn ngữ hàng đầu

Biểu đồ hiển thị số lượng quốc gia theo ngôn ngữ chính thức hàng đầu:

Tiếng Pháp, tiếng Anh và tiếng Tây Ban Nha là những ngôn ngữ chính thức phổ biến nhất trong số 10 quốc gia hàng đầu.

Tiếng Ả Rập và tiếng Bồ Đào Nha cũng được sử dụng rộng rãi như ngôn ngữ chính thức, mỗi ngôn ngữ là ngôn ngữ chính ở khoảng 10 quốc gia.

Có một số quốc gia không có ngôn ngữ chính thức, điều này có thể cho thấy sự đa dạng về ngôn ngữ hoặc đa ngôn ngữ.

Tiếng Nga, tiếng Swahili, tiếng Ba Tư và tiếng Rumani cũng có mặt, mặc dù với số lượng ít hơn.

```python
# Xác định số lượng ngôn ngữ hàng đầu để hiển thị
top_languages_count = 10

# Lấy N ngôn ngữ hàng đầu
top_languages = df['Official language'].value_counts().nlargest(top_languages_count).index

# Lọc DataFrame chỉ chứa các quốc gia có ngôn ngữ chính thức hàng đầu
df_top_languages = df[df['Official language'].isin(top_languages)]

# Số lượng quốc gia theo Ngôn ngữ chính thức hàng đầu
plt.figure(figsize=(8, 5))
sns.countplot(data=df_top_languages, y='Official language', order=df_top_languages['Official language'].value_counts().index, palette='bright', saturation=0.8)
plt.title('Số lượng Quốc gia theo Ngôn ngữ chính thức hàng đầu')
plt.xlabel('Số lượng Quốc gia')
plt.ylabel('Ngôn ngữ chính thức')
plt.tight_layout()

# Thêm chú thích cho từng cột
for index, value in enumerate(df_top_languages['Official language'].value_counts()):
    plt.text(value + 1, index, f'{value}', va='center')

plt.show()
```

20| Phân phối mã tiền tệ

Phân phối tiền tệ bao gồm EUR (11,8%), XOF (6,4%), XAF (1,8%), USD (1,8%), v.v. và các loại khác (72,7%).

```python
# Lấy N mã tiền tệ hàng đầu (ví dụ, top 10)
top_n_currency = 10

# Đếm số lượng mã tiền tệ và lấy N mã hàng đầu
currency_counts = df['Currency-Code'].value_counts().head(top_n_currency)

# Tính tổng số lượng mã còn lại
remaining_counts = df['Currency-Code'].value_counts().sum() - currency_counts.sum()

# Tạo một Series mới cho các mã tiền tệ hàng đầu và thêm danh mục "Khác"
currency_counts = currency_counts.append(pd.Series([remaining_counts], index=['Other']))

# Vẽ biểu đồ tròn
color = ['red', 'green', 'blue', 'cyan','orange','grey']
plt.pie(currency_counts, labels=currency_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("rainbow_r"))
plt.title('Phân phối Mã tiền tệ')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
